```python
#using PyPDF2 library to extract the text from PDF

import PyPDF2

# Get the filename from the uploaded dictionary
filename = list(uploaded.keys())[0]
print(filename)

```

```python
def clean_text(text):
    # Remove extra whitespace
    text = " ".join(text.split())

    # Remove special characters
    ext = re.sub(r"[^a-zA-Z0-9@./,_!?'\s-]", "", text)

    return text

def extract_text_from_pdf(file_path):
    text = ""
    # Open the file in read binary ('rb') mode for PDF files
    with open(file_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            text += page.extract_text()
    # Clean the extracted text
    text = text.lower()
    cleaned_text = clean_text(text)

    return cleaned_text


# Test the function with a PDF filename
pdf_text = extract_text_from_pdf(filename)
print(pdf_text)


# Split the cleaned text into words
words = pdf_text.split()

# Use a set to get unique words
unique_words = set(words)
print(unique_words)
# Get the vocabulary size (total unique words)
vocab_size = len(unique_words)

print(f"Vocabulary size: {vocab_size}")
```

Now Building  NLP-based Question-Answering (Q&A) system.


```python
from transformers import pipeline

# Load pre-trained DistilBERT model for Question-Answering
qa_pipeline = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad", tokenizer="distilbert-base-uncased-distilled-squad")

def answer_question(question, context):
    # Get the answer from the model
    answer = qa_pipeline({
        'question': question,
        'context': context
    })
    return answer['answer']
# Context: the text extracted from your PDF
context = pdf_text  # Replace this with your actual extracted text

print("Welcome to the Q&A system!")
print("You can start asking questions related to the PDF content.")

while True:
    # Get user input (question)
    question = input("Enter your question (or type 'exit' to quit): ")

    # Exit condition
    if question.lower() == 'exit':
        print("Goodbye!")
        break

    # Get the answer from the model
    answer = answer_question(question, context)

    # Print the answer
    print(f"Answer: {answer}")
```
